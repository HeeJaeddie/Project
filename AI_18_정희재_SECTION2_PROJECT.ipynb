```python
# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ìï®Ïàò ÏÑ§Ï†ï
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lr= Q1-(1.5 * IQR)
    ur= Q3+(1.5 * IQR)
    return lr, ur
```

```python
# Ïù¥ÏÉÅÏπò Ï†úÍ±∞
for column in df.columns:
    if df[column].dtype != 'object': 
        lr,ur=remove_outlier(df[column])
        df[column]=np.where(df[column]>ur,ur,df[column])
        df[column]=np.where(df[column]<lr,lr,df[column])
```

- PDP(Partial Dependence Plot)Îäî Î™®Îç∏ÏóêÏÑú ÌïòÎÇòÏùò ÌäπÏÑ±(feature)Ïù¥ Î≥ÄÌï† Îïå Í∑∏Ïóê Îî∞Î•∏ ÏòàÏ∏°Í∞íÏùò Î≥ÄÌôîÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Î∞©Î≤ïÏûÖÎãàÎã§. Ï¶â, ÌäπÏ†ï Î≥ÄÏàòÍ∞Ä Î™®Îç∏ ÏòàÏ∏°Ïóê ÎØ∏ÏπòÎäî ÏòÅÌñ•Î†•ÏùÑ Ïù¥Ìï¥ÌïòÍ∏∞ ÏúÑÌïú Î∞©Î≤ï Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

```python
from pdpbox.pdp import pdp_isolate, pdp_plot
from pdpbox.pdp import pdp_interact, pdp_interact_plot

X_val3_encoded = pipe_xgb[0].transform(X_val3)

# Tenure
isolated = pdp_isolate(
    model=pipe_xgb[1],
    dataset=X_val3_encoded,
    model_features=X_val3.columns,
    feature='Tenure',
    grid_type="percentile",  
    num_grid_points=10
)
pdp_plot(isolated, feature_name='Tenure', plot_pts_dist=True);
```
